```python
# 1-1
X = df[['calory', 'breakfast', 'lunch', 'dinner', 'exercise']]
Y = df[['body_shape']]

x_std = StandardScaler().fit_transform(X)
x_std
```

```python
# 1-2
features = x_std.T
covariance_matrix = np.cov(features)
print(covariance_matrix)

''' 공분산 행렬 구현
mean_X = np.mean(features, axis=0)

centered_data = features - mean_X

cov_matrix = np.dot(centered_data.T, centered_data) / (features.shape[0] - 1)
'''

eig_vals, eig_vecs = np.linalg.eig(covariance_matrix)
print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
# 1-3
projected_X = x_std.dot(eig_vecs.T[0]) / np.linalg.norm(eig_vecs.T[0])
projected_X
```

```python
# 1-4
result = pd.DataFrame(projected_X, columns=['PC1'])
result['y-axis'] = 0.0
result['label'] = Y

result.head(10)

sns.scatterplot(x='PC1', y='y-axis', data=result, hue='label', s=50)
plt.title('PCA result')
```
